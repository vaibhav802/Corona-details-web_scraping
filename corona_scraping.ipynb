```python
%%writefile corona.py
def Corona_details():
    """This function gives the deatails of  Corona  in India
    """
    import requests
    from bs4 import BeautifulSoup as bs
    url = 'https://www.mygov.in/corona-data/covid19-statewise-status/'
    page = requests.get(url)
    obj = bs(page.content, 'html')
    get_table_state = obj.find_all('div', attrs={'class':'field field-name-field-select-state field-type-list-text field-label-above'})
    state = []
    for i in get_table_state:
        n = (i.text).split('\xa0')[-1]
        state.append(n)
    get_table_confirmed = obj.find_all('div', attrs={'field field-name-field-total-confirmed-indians field-type-number-integer field-label-above'})
    confirmed = []
    for no in get_table_confirmed:
        n = int((no.text).split('\xa0')[-1]) 
        confirmed.append(n)
    get_table_discharge =obj.find_all('div', attrs='field field-name-field-cured field-type-number-integer field-label-above')
    discharge = []
    for no in get_table_discharge:
        n = int((no.text).split('\xa0')[-1])
        discharge.append(n)
    get_table_death = obj.find_all('div', attrs='field field-name-field-deaths field-type-number-integer field-label-above')
    death = []
    for no in get_table_death:
        n = int((no.text).split('\xa0')[-1])
        death.append(n)

    def main():
        """This function gives the Menu to be slected
        """
        list1 = ['Most affected State', \
             'Least affected State', \
             'Top 5 most confirmed cases - State & Num.',\
             'Top 5 least confirmed cases - State & Num.',
             'Top 5 most discharge cases - State & Num.', \
             'Top 5 least discharge cases - State & Num.',\
             'Top 5 most death cases - State & Num.',\
             'Top 5 least death cases - State & Num','Exit']
        print('\n\nList of State Wise Confirmed, Discharge & Death numbers due to Corona \n\n'.center(120))
        for i in range(len(list1)):
            print(f'{i+1}. {list1[i]}')
        Choice = int(input('\n\nEnter a Number from the above options : '))
        if Choice == 1:
            Most_Affected()
        elif Choice == 2:
            Least_Affected()
        elif Choice == 3:
            Top_5_confirmedcases_state()
        elif Choice == 4:
            least_5_confirmedcases_state()
        elif Choice == 5:
            Top_5_dischargecases_state()
        elif Choice == 6:
            least_5_dischargecases_state()
        elif Choice == 7:
            Top_5_deathcases_state()
        elif Choice ==8:
            least_5_deathcases_state()
        elif Choice == 9:
            print('\nStay Home Stay Safe'.center(120))
        else:
            print('\nEnter a valid number from the above\n')
            main()
    
    def Most_Affected(S =state.copy(), N = 1,C=confirmed.copy() ,Di =discharge.copy(), D = death.copy()):
        """This Function gives Most_Affected State
        """
        Final_state1_list = []
        Final_state2_list = []
        Final_state3_list = []
        Final_confirmed_list = []
        Final_discharge_list = []
        Final_death_list = []
        for i in range(N):
            con = 0
            for j in range(len(C)):
                if C[j] >= con:
                    con = C[j]
                    State1 = S[j]
            C.remove(con)
            Final_state1_list.append(State1)
            Final_confirmed_list.append(con)
        for i in range(N):
            Dis = 0
            for j in range(len(Di)):
                if Di[j] >= Dis:
                    Dis = Di[j]
                    State2 = S[j]
            Di.remove(Dis)
            Final_state2_list.append(State2)
            Final_discharge_list.append(Dis)
        for i in range(N):
            De = 0
            for j in range(len(D)):
                if D[j] >= De:
                    De = D[j]
                    State3 = S[j]
            D.remove(De)
            Final_state3_list.append(State3)
            Final_death_list.append(De)

        print(f'\nMost confirmed Cases -State & Num.: {Final_state1_list} , {Final_confirmed_list}')
        print(f'\nMost discharged Cases -State & Num.: {Final_state2_list} , {Final_discharge_list}')
        print(f'\nMost death Cases -State & Num.: {Final_state3_list} , {Final_death_list}\n')
        main()
    
    def Least_Affected(State =state.copy(), Number = 1,C=confirmed.copy() ,Di =discharge.copy(), D = death.copy() ):
        Final_state1_list = []
        Final_state2_list = []
        Final_state3_list = []
        Final_confirmed_list = []
        Final_discharge_list = []
        Final_death_list = []
        for i in range(Number):
            con = C[0]
            for j in range(len(C)):
                if C[j] <= con:
                    con = C[j]
                    State1 = State[j]
            C.remove(con)
            Final_state1_list.append(State1)
            Final_confirmed_list.append(con)
        for i in range(Number):
            Dis = Di[0]
            for j in range(len(Di)):
                if Di[j] <= Dis:
                    Dis = Di[j]
                    State2 = State[j]
            Di.remove(Dis)
            Final_state2_list.append(State2)
            Final_discharge_list.append(Dis)
        for i in range(Number):
            De = D[0]
            for j in range(len(D)):
                if D[j] <= De:
                    De = D[j]
                    State3 = State[j]
            D.remove(De)
            Final_state3_list.append(State3)
            Final_death_list.append(De)

        print(f'\nLeast confirmed Cases -State & Num.:  {Final_state1_list} , {Final_confirmed_list}')
        print(f'\nLeast discharged Cases -State & Num.:  {Final_state2_list} , {Final_discharge_list}')
        print(f'\nLeast death Cases -State & Num.:  {Final_state3_list} , {Final_death_list}\n')
        main()

    def Top_5_confirmedcases_state(State =state.copy(), Number = 5, Confirmed=confirmed.copy()):
        Final_state_list = []
        Final_confirmed_list = []
        for i in range(Number):
            max1 = 0
            for j in range(len(Confirmed)):
                if Confirmed[j] >= max1:
                    max1 = Confirmed[j]
                    State1 = State[j]
            Confirmed.remove(max1)
            State.remove(State1)
            Final_state_list.append(State1)
            Final_confirmed_list.append(max1)
        print('\n\n Top 5 confirmed cases - State & Num.'.center(120))
        print('\n', dict(zip(Final_state_list,Final_confirmed_list)))
        main()

    
    def Top_5_dischargecases_state(State = state.copy(), Number = 5, Confirmed = discharge.copy()):
        Final_state_list = []
        Final_discharge_list = []
        for i in range(Number):
            max1 = 0
            for j in range(len(Confirmed)):
                if Confirmed[j] >= max1:
                    max1 = Confirmed[j]
                    State1 = State[j]
            Confirmed.remove(max1)
            State.remove(State1)
            Final_state_list.append(State1)
            Final_discharge_list.append(max1)
        print('\n\n Top 5 dischargedcases - State & Num.'.center(120))
        print('\n', dict(zip(Final_state_list,Final_discharge_list)))
        main()
    
    
    def least_5_deathcases_state(State = state.copy(), Number = 5, Confirmed = death.copy()):
        Final_state_list = []
        Final_death_list = []
        for i in range(Number):
            max1 = Confirmed[0]
            for j in range(len(Confirmed)):
                if Confirmed[j] <= max1:
                    max1 = Confirmed[j]
                    State1 = State[j]
            Confirmed.remove(max1)
            State.remove(State1)
            Final_state_list.append(State1)
            Final_death_list.append(max1)
        print('\n\n Least 5 death cases - State & Num.'.center(120))
        print('\n' ,dict(zip(Final_state_list,Final_death_list)))
        main()

    
    def least_5_dischargecases_state(State = state.copy(), Number = 5, Confirmed = discharge.copy()):
        Final_state_list = []
        Final_discharge_list = []
        for i in range(Number):
            max1 = Confirmed[0]
            for j in range(len(Confirmed)):
                if Confirmed[j] <= max1:
                    max1 = Confirmed[j]
                    State1 = State[j]
            Confirmed.remove(max1)
            State.remove(State1)
            Final_state_list.append(State1)
            Final_discharge_list.append(max1)
        print('\n\n Least 5 discharged cases - State & Num.'.center(120))
        print('\n',dict(zip(Final_state_list,Final_discharge_list)))
        main()

    
    def least_5_confirmedcases_state(State = state.copy(), Number = 5, Confirmed = confirmed.copy()):
        Final_state_list = []
        Final_confirmed_list = []
        for i in range(Number):
            max1 = Confirmed[0]
            for j in range(len(Confirmed)):
                if Confirmed[j] <= max1:
                    max1 = Confirmed[j]
                    State1 = State[j]
            Confirmed.remove(max1)
            State.remove(State1)
            Final_state_list.append(State1)
            Final_confirmed_list.append(max1)
        print('\n\n Least 5 confirmed cases - State & Num.'.center(120))
        print('\n', dict(zip(Final_state_list,Final_confirmed_list)))
        main()
    
    def Top_5_deathcases_state(state_name = state.copy(), number = 5, confirm = death.copy()):
        Final_state_list = []
        Final_death_list = []
        for i in range(number):
            max1 = 0
            for j in range(len(confirm)):
                if confirm[j] >= max1:
                    max1 = confirm[j]
                    State1 = state_name[j]
            confirm.remove(max1)
            state_name.remove(State1)
            Final_state_list.append(State1)
            Final_death_list.append(max1)
        print('\n\n Top 5 death cases - State & Num.'.center(120))
        print('\n',dict(zip(Final_state_list,Final_death_list)))
        main()
    main()
    
        
    
```
